## Case report.csv

```python
folder_path = 'C://Users/ACER/Desktop/DATH/'
```

```python
df_case_report = pd.read_csv(folder_path + 'Cases Report.csv')
df_case_report.info()
```

```python
print(" \nTotal missing value with attribute : \n")
print(df_case_report.isnull().sum())
```

```python
df_case_report[df_case_report.duplicated(keep=False)]
```

#### Cleaning data

```python
df_case_report['Gender'] = df_case_report['Gender'].str.capitalize()
df_case_report['OutbreakRelated'] = df_case_report['OutbreakRelated'].fillna('Unknown')
date_col = ['SpecimenDate', 'CaseReported Date', 'TestReported Date', 'AccurateEpisode Dt']

df_case_report[date_col] = df_case_report[date_col].apply(pd.to_datetime)
```

```python
df_case_report['OutbreakRelated'].unique()
```

```python
df_clean_case_report = df_case_report
df_clean_case_report
```

```python
print(" \nTotal missing value with attribute : \n\n",
      df_clean_case_report.isnull().sum())
```

## vaccines_by_age_phu

```python
df_vaccines_by_age_phu = pd.read_csv(folder_path + '/vaccines_by_age_phu.csv')

df_vaccines_by_age_phu.info()
print("\nTotal missing value with attribute : \n") 
print(df_vaccines_by_age_phu.isnull().sum())
```

```python
df_vaccines_by_age_phu
```

```python
df_vaccines_by_age_phu_clean = df_vaccines_by_age_phu
df_vaccines_by_age_phu_clean
```

```python
print(" \nTotal missing value with attribute : \n")
print(df_clean_vaccines_by_age_phu.isnull().sum())
```
